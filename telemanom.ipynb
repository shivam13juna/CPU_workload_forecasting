```python
train_dataset = np.expand_dims(train_dataset, axis = 1)
test_dataset = np.expand_dims(test_dataset, axis = 1)
```

```python
cbs = [EarlyStopping(monitor='val_loss', patience=10,  verbose=2)]
        
model = Sequential()

model.add(LSTM(80,
    input_shape=(None, 15),
    return_sequences=True))
model.add(Dropout(0.5))

model.add(LSTM(
    80,
    return_sequences=False))
model.add(Dropout(0.5))

model.add(Dense(10))
model.add(Activation("linear"))

model.compile(loss='mse', optimizer=tensorflow.keras.optimizers.Adam(0.001)) 

# model.save(os.path.join("data", anom['run_id'], "models", anom["chan_id"] + ".h5"))
```

```python
model.fit(train_dataset, trainy, batch_size=70, epochs=10000, 
    validation_data=[test_dataset, testy], callbacks=cbs, verbose=True)
```
